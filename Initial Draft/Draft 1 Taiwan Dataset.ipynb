```python
# Select a feature to analyze
col_to_analyze = 'bill_amt1'

# Calculate and print its original skewness
original_skewness = df[col_to_analyze].skew()
print(f"--- Outlier Analysis for '{col_to_analyze}' ---")
print(f"Original Skewness: {original_skewness:.4f}")

sns.histplot(df[col_to_analyze], kde=True)
plt.title(f'Original Distribution of {col_to_analyze}')
plt.show()

# Observation: The initial skewness is high and positive, confirming a long right tail caused by outliers, which we also saw in the box plots.
```

```python
# Create a copy of the DataFrame for testing
df_test = df.copy()

# --- Identify Outliers using the IQR method ---
Q1 = df_test[col_to_analyze].quantile(0.25)
Q3 = df_test[col_to_analyze].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier_mask = (df_test[col_to_analyze] < lower_bound) | (df_test[col_to_analyze] > upper_bound)

# --- Option A: Replace Outliers with the Median ---
median_val = df_test[col_to_analyze].median()
df_test['median_imp'] = df_test[col_to_analyze].where(~outlier_mask, median_val)
median_imp_skew = df_test['median_imp'].skew()

# --- Option B: Capping (Winsorizing) at 1st and 99th Percentiles ---
p1 = df_test[col_to_analyze].quantile(0.01)
p99 = df_test[col_to_analyze].quantile(0.99)
df_test['capped'] = np.clip(df_test[col_to_analyze], p1, p99)
capped_skew = df_test['capped'].skew()

# --- Option C: Log Transformation ---
# We shift the data to ensure all values are positive before applying the log
df_test['log_transform'] = np.log1p(df_test[col_to_analyze] - df_test[col_to_analyze].min())
log_skew = df_test['log_transform'].skew()

# Create a summary DataFrame
results = pd.DataFrame({
    'Method': ['Original Data', 'Replace with Median', 'Capping (Winsorizing)', 'Log Transform'],
    'Skewness': [original_skewness, median_imp_skew, capped_skew, log_skew]
}).sort_values(by='Skewness', key=abs)

print("--- Skewness Comparison ---")
print(results)

"""Analysis and Decision:

Replacing with Median and Log Transform both created new problems (heavy distortion or overcorrection to a negative skew), as we saw in our previous detailed analysis.

Capping (Winsorizing) is the clear winner. It dramatically reduced skewness to 0.07, which is very close to symmetrical, without artificially distorting the data's underlying distribution.

Conclusion: The best and most robust option is Capping."""
```

```python
print("Applying the best option (Capping) to the original DataFrame...")

# Define all numerical columns that need capping
cols_to_cap = ['bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6',
               'pay_amt1', 'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6',
               'limit_bal']

for col in cols_to_cap:
    p1 = df[col].quantile(0.01)
    p99 = df[col].quantile(0.99)
    
    # Create a NEW column with a '_capped' suffix
    df[f'{col}_capped'] = np.clip(df[col], p1, p99)
    

# --- Verify by checking the first few rows ---
display(df[['limit_bal', 'limit_bal_capped', 'bill_amt1', 'bill_amt1_capped']].head())
```

## 2. Data Preparation & Feature Engineering

```python
# Data Preparation for X-FuzzyScore Framework

# Select features for modeling
feature_cols = [
    'limit_bal_capped', 'sex', 'education', 'marriage', 'age',
    'pay_1', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6',
    'bill_amt1_capped', 'bill_amt2_capped', 'bill_amt3_capped', 
    'bill_amt4_capped', 'bill_amt5_capped', 'bill_amt6_capped',
    'pay_amt1_capped', 'pay_amt2_capped', 'pay_amt3_capped',
    'pay_amt4_capped', 'pay_amt5_capped', 'pay_amt6_capped'
]

# Prepare feature matrix and target
X = df[feature_cols].copy()
y = df['default_payment'].copy()

print(f"Dataset: {X.shape[0]} samples, {X.shape[1]} features")
print(f"Default rate: {y.mean()*100:.2f}%")

# Train-Test Split (80-20)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert to DataFrame for tracking
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

print(f"Training: {len(X_train)} | Testing: {len(X_test)}")
print("✓ Data preparation complete")
```

## 3. Baseline Machine Learning Models

```python
# Baseline Machine Learning Models

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    """Train and evaluate classification model"""
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None
    
    metrics = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else None
    }
    
    print(f"{model_name}: ROC-AUC = {metrics['ROC-AUC']:.4f}, F1 = {metrics['F1-Score']:.4f}")
    
    return {
        'model': model,
        'name': model_name,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba,
        'metrics': metrics
    }

print("Training baseline models...\n")
baseline_results = {}

# Logistic Regression
lr_model = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
baseline_results['Logistic Regression'] = evaluate_model(
    lr_model, X_train_scaled, X_test_scaled, y_train, y_test, 'Logistic Regression'
)

# Random Forest
rf_model = RandomForestClassifier(
    n_estimators=100, max_depth=10, random_state=42, 
    class_weight='balanced', n_jobs=-1
)
baseline_results['Random Forest'] = evaluate_model(
    rf_model, X_train_scaled, X_test_scaled, y_train, y_test, 'Random Forest'
)

print("\n✓ Baseline models trained")
```

```python
# Train Ensemble Models (XGBoost & LightGBM)

print("Training ensemble models...\n")

# XGBoost
xgb_model = xgb.XGBClassifier(
    n_estimators=100, max_depth=6, learning_rate=0.1,
    random_state=42, eval_metric='logloss',
    scale_pos_weight=len(y_train[y_train==0]) / len(y_train[y_train==1])
)
baseline_results['XGBoost'] = evaluate_model(
    xgb_model, X_train_scaled, X_test_scaled, y_train, y_test, 'XGBoost'
)

# LightGBM
lgb_model = lgb.LGBMClassifier(
    n_estimators=100, max_depth=6, learning_rate=0.1,
    random_state=42, class_weight='balanced', verbosity=-1
)
baseline_results['LightGBM'] = evaluate_model(
    lgb_model, X_train_scaled, X_test_scaled, y_train, y_test, 'LightGBM'
)

print("\n✓ Ensemble models trained")
```

```python
# Baseline Model Comparison

# Create comparison DataFrame
comparison_data = []
for model_name, results in baseline_results.items():
    row = {'Model': model_name}
    row.update(results['metrics'])
    comparison_data.append(row)

baseline_comparison = pd.DataFrame(comparison_data).sort_values('ROC-AUC', ascending=False)

print("Baseline Model Performance:\n")
display(baseline_comparison.style.format({
    'Accuracy': '{:.4f}', 'Precision': '{:.4f}', 'Recall': '{:.4f}',
    'F1-Score': '{:.4f}', 'ROC-AUC': '{:.4f}'
}).background_gradient(subset=['ROC-AUC'], cmap='YlGn'))

best_model_name = baseline_comparison.iloc[0]['Model']
best_model_result = baseline_results[best_model_name]
print(f"\nBest Model: {best_model_name} (ROC-AUC: {best_model_result['metrics']['ROC-AUC']:.4f})")

# Visualization: Model Performance
fig = go.Figure()
metrics_to_plot = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']

for metric in metrics_to_plot:
    fig.add_trace(go.Bar(
        name=metric,
        x=baseline_comparison['Model'],
        y=baseline_comparison[metric],
        text=baseline_comparison[metric].round(3),
        textposition='auto',
    ))

fig.update_layout(
    title='Baseline Model Performance Comparison',
    xaxis_title='Model', yaxis_title='Score',
    barmode='group', height=500, template='plotly_white'
)
fig.show()
```

## 4. Fuzzy Logic System (FIS)

Implement fuzzy inference system using scikit-fuzzy with interpretable linguistic rules.

```python
# Fuzzy Inference System for Interpretable Risk Scoring

# Define Fuzzy Variables
credit_limit = ctrl.Antecedent(np.arange(0, 1.01, 0.01), 'credit_limit')
credit_limit['low'] = fuzz.trimf(credit_limit.universe, [0, 0, 0.4])
credit_limit['medium'] = fuzz.trimf(credit_limit.universe, [0.2, 0.5, 0.8])
credit_limit['high'] = fuzz.trimf(credit_limit.universe, [0.6, 1, 1])

payment_status = ctrl.Antecedent(np.arange(0, 1.01, 0.01), 'payment_status')
payment_status['good'] = fuzz.trimf(payment_status.universe, [0, 0, 0.3])
payment_status['fair'] = fuzz.trimf(payment_status.universe, [0.2, 0.5, 0.8])
payment_status['poor'] = fuzz.trimf(payment_status.universe, [0.7, 1, 1])

bill_ratio = ctrl.Antecedent(np.arange(0, 1.01, 0.01), 'bill_ratio')
bill_ratio['low'] = fuzz.trimf(bill_ratio.universe, [0, 0, 0.4])
bill_ratio['medium'] = fuzz.trimf(bill_ratio.universe, [0.3, 0.6, 0.9])
bill_ratio['high'] = fuzz.trimf(bill_ratio.universe, [0.8, 1, 1])

age_input = ctrl.Antecedent(np.arange(0, 1.01, 0.01), 'age')
age_input['young'] = fuzz.trimf(age_input.universe, [0, 0, 0.4])
age_input['middle'] = fuzz.trimf(age_input.universe, [0.3, 0.5, 0.7])
age_input['senior'] = fuzz.trimf(age_input.universe, [0.6, 1, 1])

risk_score = ctrl.Consequent(np.arange(0, 101, 1), 'risk_score')
risk_score['low'] = fuzz.trimf(risk_score.universe, [0, 0, 40])
risk_score['medium'] = fuzz.trimf(risk_score.universe, [30, 50, 70])
risk_score['high'] = fuzz.trimf(risk_score.universe, [60, 100, 100])

# Define Fuzzy Rules
rule1 = ctrl.Rule(credit_limit['low'] & payment_status['poor'], risk_score['high'])
rule2 = ctrl.Rule(credit_limit['low'] & bill_ratio['high'], risk_score['high'])
rule3 = ctrl.Rule(payment_status['poor'] & bill_ratio['high'], risk_score['high'])
rule4 = ctrl.Rule(credit_limit['high'] & payment_status['good'], risk_score['low'])
rule5 = ctrl.Rule(credit_limit['medium'] & payment_status['fair'], risk_score['medium'])
rule6 = ctrl.Rule(age_input['young'] & payment_status['poor'], risk_score['high'])
rule7 = ctrl.Rule(age_input['senior'] & payment_status['good'], risk_score['low'])
rule8 = ctrl.Rule(bill_ratio['low'] & payment_status['good'], risk_score['low'])

# Create control system
risk_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8])
risk_simulation = ctrl.ControlSystemSimulation(risk_ctrl)

print(f"✓ Fuzzy Inference System created with 8 rules\n")

# Visualize membership functions
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Fuzzy Membership Functions', fontsize=16, fontweight='bold')

credit_limit.view(ax=axes[0, 0])
axes[0, 0].set_title('Credit Limit', fontsize=12)

payment_status.view(ax=axes[0, 1])
axes[0, 1].set_title('Payment Status', fontsize=12)

bill_ratio.view(ax=axes[0, 2])
axes[0, 2].set_title('Bill Amount Ratio', fontsize=12)

age_input.view(ax=axes[1, 0])
axes[1, 0].set_title('Age', fontsize=12)

risk_score.view(ax=axes[1, 1])
axes[1, 1].set_title('Risk Score (Output)', fontsize=12)

axes[1, 2].axis('off')

plt.tight_layout()
plt.show()

fuzzy_available = True
```

```python
# Compute Fuzzy Risk Scores

def normalize_feature(values, min_val=None, max_val=None):
    """Normalize feature to [0, 1] range"""
    if min_val is None:
        min_val = values.min()
    if max_val is None:
        max_val = values.max()
    if max_val == min_val:
        return np.zeros_like(values)
    return (values - min_val) / (max_val - min_val)

def compute_fuzzy_score(row, simulation):
    """Compute fuzzy risk score for a single data point"""
    try:
        simulation.input['credit_limit'] = row['credit_limit_norm']
        simulation.input['payment_status'] = row['payment_status_norm']
        simulation.input['bill_ratio'] = row['bill_ratio_norm']
        simulation.input['age'] = row['age_norm']
        simulation.compute()
        return simulation.output['risk_score']
    except:
        return 50.0

# Prepare features for fuzzy computation
limit_col = 'limit_bal_capped' if 'limit_bal_capped' in X_train.columns else 'limit_bal'
bill_col = 'bill_amt1_capped' if 'bill_amt1_capped' in X_train.columns else 'bill_amt1'
pay_cols = ['pay_1', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']

# Training set
X_train_fuzzy = X_train.copy()
X_train_fuzzy['credit_limit_norm'] = normalize_feature(X_train[limit_col])
X_train_fuzzy['payment_status_norm'] = normalize_feature(X_train[pay_cols].mean(axis=1), 0, 8)
X_train_fuzzy['bill_ratio_norm'] = np.where(
    X_train[limit_col] > 0,
    normalize_feature(np.clip(X_train[bill_col] / X_train[limit_col], 0, 1)),
    0
)
X_train_fuzzy['age_norm'] = normalize_feature(X_train['age'], 20, 80)

# Test set
X_test_fuzzy = X_test.copy()
X_test_fuzzy['credit_limit_norm'] = normalize_feature(X_test[limit_col])
X_test_fuzzy['payment_status_norm'] = normalize_feature(X_test[pay_cols].mean(axis=1), 0, 8)
X_test_fuzzy['bill_ratio_norm'] = np.where(
    X_test[limit_col] > 0,
    normalize_feature(np.clip(X_test[bill_col] / X_test[limit_col], 0, 1)),
    0
)
X_test_fuzzy['age_norm'] = normalize_feature(X_test['age'], 20, 80)

# Compute fuzzy scores
print("Computing fuzzy risk scores...")
fuzzy_scores_train = np.array([
    compute_fuzzy_score(row, risk_simulation) 
    for _, row in X_train_fuzzy.iterrows()
]) / 100.0

fuzzy_scores_test = np.array([
    compute_fuzzy_score(row, risk_simulation) 
    for _, row in X_test_fuzzy.iterrows()
]) / 100.0

print(f"✓ Fuzzy scores computed: Range [{fuzzy_scores_test.min():.3f}, {fuzzy_scores_test.max():.3f}]\n")

# Visualize fuzzy score distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(fuzzy_scores_train, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
axes[0].set_title('Fuzzy Risk Score Distribution', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Fuzzy Risk Score')
axes[0].set_ylabel('Frequency')
axes[0].axvline(fuzzy_scores_train.mean(), color='red', linestyle='--', 
                label=f'Mean: {fuzzy_scores_train.mean():.3f}')
axes[0].legend()

axes[1].boxplot([fuzzy_scores_train[y_train==0], fuzzy_scores_train[y_train==1]], 
                labels=['No Default', 'Default'])
axes[1].set_title('Fuzzy Risk Score by Default Status', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Fuzzy Risk Score')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 5. X-FuzzyScore Hybrid Models

Combine fuzzy risk scores with ML models for enhanced interpretability.

```python
# X-FuzzyScore Hybrid Models

# Add fuzzy score as new feature
X_train_hybrid = np.column_stack([X_train_scaled, fuzzy_scores_train])
X_test_hybrid = np.column_stack([X_test_scaled, fuzzy_scores_test])
hybrid_feature_names = list(X_train.columns) + ['fuzzy_risk_score']

print(f"Enhanced features: {X_train_hybrid.shape[1]} (original: {X_train_scaled.shape[1]} + fuzzy: 1)\n")
print("Training hybrid models...\n")

hybrid_results = {}

# Hybrid Logistic Regression
lr_hybrid = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
hybrid_results['LR + Fuzzy'] = evaluate_model(
    lr_hybrid, X_train_hybrid, X_test_hybrid, y_train, y_test, 'LR + Fuzzy'
)

# Hybrid Random Forest
rf_hybrid = RandomForestClassifier(
    n_estimators=100, max_depth=10, random_state=42, 
    class_weight='balanced', n_jobs=-1
)
hybrid_results['RF + Fuzzy'] = evaluate_model(
    rf_hybrid, X_train_hybrid, X_test_hybrid, y_train, y_test, 'RF + Fuzzy'
)

# Hybrid XGBoost
xgb_hybrid = xgb.XGBClassifier(
    n_estimators=100, max_depth=6, learning_rate=0.1,
    random_state=42, eval_metric='logloss',
    scale_pos_weight=len(y_train[y_train==0]) / len(y_train[y_train==1])
)
hybrid_results['XGBoost + Fuzzy'] = evaluate_model(
    xgb_hybrid, X_train_hybrid, X_test_hybrid, y_train, y_test, 'XGBoost + Fuzzy'
)

# Hybrid LightGBM
lgb_hybrid = lgb.LGBMClassifier(
    n_estimators=100, max_depth=6, learning_rate=0.1,
    random_state=42, class_weight='balanced', verbosity=-1
)
hybrid_results['LightGBM + Fuzzy'] = evaluate_model(
    lgb_hybrid, X_train_hybrid, X_test_hybrid, y_train, y_test, 'LightGBM + Fuzzy'
)

print("\n✓ Hybrid models trained")
```

## 6. Explainability Layer (XAI)

SHAP and LIME analysis for transparent model predictions.

```python
# SHAP Explainability Analysis (Simplified & Robust)

print("Initializing SHAP analysis...\n")

# Use best hybrid model
best_hybrid_name = list(hybrid_results.keys())[0]
best_hybrid_model = hybrid_results[best_hybrid_name]['model']
print(f"Analyzing: {best_hybrid_name}")

# Use smaller samples for faster computation
sample_size = min(200, X_train_hybrid.shape[0])
test_sample_size = min(100, X_test_hybrid.shape[0])

X_train_sample = X_train_hybrid[:sample_size]
X_test_sample = X_test_hybrid[:test_sample_size]

# Create SHAP explainer (optimized for tree models)
if 'XGBoost' in best_hybrid_name or 'LightGBM' in best_hybrid_name:
    explainer = shap.TreeExplainer(best_hybrid_model)
    print("Using TreeExplainer (fast)\n")
else:
    explainer = shap.KernelExplainer(
        lambda x: best_hybrid_model.predict_proba(x)[:, 1],  # Only return class 1 probabilities
        shap.sample(X_train_sample, min(50, len(X_train_sample)))
    )
    print("Using KernelExplainer\n")

# Compute SHAP values
print("Computing SHAP values...")
shap_values = explainer.shap_values(X_test_sample)

# Handle different SHAP value formats
if isinstance(shap_values, list):
    # Tree models return list [class0, class1]
    shap_values_2d = shap_values[1]
elif shap_values.ndim == 3:
    # Multi-output format: (samples, classes, features)
    shap_values_2d = shap_values[:, 1, :]
else:
    # Already 2D: (samples, features)
    shap_values_2d = shap_values

print(f"✓ SHAP values computed: {shap_values_2d.shape}")
print(f"  Samples: {shap_values_2d.shape[0]}, Features: {shap_values_2d.shape[1]}\n")

# Verify shapes match
if shap_values_2d.shape[1] != len(hybrid_feature_names):
    print(f"⚠ Shape mismatch detected:")
    print(f"  SHAP values features: {shap_values_2d.shape[1]}")
    print(f"  Feature names: {len(hybrid_feature_names)}")
    print("  Skipping visualization and continuing with feature importance calculation...\n")
else:
    # Visualization: Summary Plot (Feature Importance)
    try:
        plt.figure(figsize=(10, 6))
        shap.summary_plot(
            shap_values_2d, 
            X_test_sample, 
            feature_names=hybrid_feature_names,
            plot_type='bar',
            show=False,
            max_display=15
        )
        plt.title('SHAP Feature Importance (Top 15)', fontsize=12, fontweight='bold')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Warning: Could not create summary plot: {e}")

# Feature Importance Table
if shap_values_2d.shape[1] == len(hybrid_feature_names):
    feature_importance = pd.DataFrame({
        'Feature': hybrid_feature_names,
        'Mean |SHAP|': np.abs(shap_values_2d).mean(axis=0)
    }).sort_values('Mean |SHAP|', ascending=False)

    print("Top 10 Most Important Features:\n")
    display(feature_importance.head(10).style.format({'Mean |SHAP|': '{:.4f}'})
           .background_gradient(subset=['Mean |SHAP|'], cmap='YlOrRd'))

    # Check fuzzy score importance
    if 'fuzzy_risk_score' in feature_importance['Feature'].values:
        fuzzy_rank = feature_importance[feature_importance['Feature'] == 'fuzzy_risk_score'].index[0] + 1
        fuzzy_shap = feature_importance[feature_importance['Feature'] == 'fuzzy_risk_score']['Mean |SHAP|'].values[0]
        print(f"\n✓ Fuzzy Risk Score: Rank #{fuzzy_rank}, Mean |SHAP| = {fuzzy_shap:.4f}")
    
    shap_available = True
    print("\n✓ SHAP analysis complete")
else:
    # Create simple feature importance from model coefficients
    print("Using model coefficients for feature importance (SHAP unavailable):\n")
    
    if hasattr(best_hybrid_model, 'coef_'):
        # Linear models (Logistic Regression)
        coefficients = np.abs(best_hybrid_model.coef_[0])
        feature_importance = pd.DataFrame({
            'Feature': hybrid_feature_names,
            'Importance': coefficients
        }).sort_values('Importance', ascending=False)
    elif hasattr(best_hybrid_model, 'feature_importances_'):
        # Tree models
        feature_importance = pd.DataFrame({
            'Feature': hybrid_feature_names,
            'Importance': best_hybrid_model.feature_importances_
        }).sort_values('Importance', ascending=False)
    else:
        print("⚠ Cannot extract feature importance from this model type")
        feature_importance = None
    
    if feature_importance is not None:
        display(feature_importance.head(10).style.format({'Importance': '{:.4f}'})
               .background_gradient(subset=['Importance'], cmap='YlOrRd'))
        
        if 'fuzzy_risk_score' in feature_importance['Feature'].values:
            fuzzy_rank = feature_importance[feature_importance['Feature'] == 'fuzzy_risk_score'].index[0] + 1
            fuzzy_imp = feature_importance[feature_importance['Feature'] == 'fuzzy_risk_score']['Importance'].values[0]
            print(f"\n✓ Fuzzy Risk Score: Rank #{fuzzy_rank}, Importance = {fuzzy_imp:.4f}")
    
    shap_available = False
    print("\n✓ Feature importance analysis complete")
```

```python
# Individual Prediction Explanations (SHAP)

# Get example indices
default_indices = np.where(y_test.iloc[:test_sample_size] == 1)[0]
no_default_indices = np.where(y_test.iloc[:test_sample_size] == 0)[0]

if len(default_indices) > 0 and len(no_default_indices) > 0:
    high_risk_idx = default_indices[0]
    low_risk_idx = no_default_indices[0]
    
    # High-risk example
    print("Example 1: High-Risk Customer (Defaulted)")
    pred_prob = best_hybrid_model.predict_proba(X_test_sample)[high_risk_idx, 1]
    print(f"Predicted default probability: {pred_prob:.3f}\n")
    
    # Create simple bar plot for top features
    top_n = 10
    shap_vals = shap_values_2d[high_risk_idx]
    top_indices = np.argsort(np.abs(shap_vals))[-top_n:][::-1]
    
    plt.figure(figsize=(10, 6))
    colors = ['red' if v > 0 else 'blue' for v in shap_vals[top_indices]]
    plt.barh(range(top_n), shap_vals[top_indices], color=colors, alpha=0.7)
    plt.yticks(range(top_n), [hybrid_feature_names[i] for i in top_indices])
    plt.xlabel('SHAP Value (Impact on Prediction)')
    plt.title('Top 10 Features - High-Risk Customer', fontsize=12, fontweight='bold')
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
    plt.tight_layout()
    plt.show()
    
    # Low-risk example
    print("\nExample 2: Low-Risk Customer (No Default)")
    pred_prob = best_hybrid_model.predict_proba(X_test_sample)[low_risk_idx, 1]
    print(f"Predicted default probability: {pred_prob:.3f}\n")
    
    # Create simple bar plot
    shap_vals = shap_values_2d[low_risk_idx]
    top_indices = np.argsort(np.abs(shap_vals))[-top_n:][::-1]
    
    plt.figure(figsize=(10, 6))
    colors = ['red' if v > 0 else 'blue' for v in shap_vals[top_indices]]
    plt.barh(range(top_n), shap_vals[top_indices], color=colors, alpha=0.7)
    plt.yticks(range(top_n), [hybrid_feature_names[i] for i in top_indices])
    plt.xlabel('SHAP Value (Impact on Prediction)')
    plt.title('Top 10 Features - Low-Risk Customer', fontsize=12, fontweight='bold')
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
    plt.tight_layout()
    plt.show()
    
    print("\n✓ Individual explanations complete")
else:
    print("Not enough samples for individual explanations")
```
